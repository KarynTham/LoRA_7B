```python
import pandas as pd
import sacrebleu

# === Step 1: Read CSV ===
df = pd.read_csv("Result_Lora.csv")

# Ground truth column
gt_col = "Trained Response"

# Prediction columns (start from 4th column â†’ index 3)
pred_cols = df.columns[3:]

# Prepare output list
results = []

for idx, row in df.iterrows():
    gt = str(row[gt_col])
    # SacreBLEU expects list of references (list of strings inside list)
    ref = [[gt]]

    row_scores = {"Row": idx, "Trained Response": gt}

    for col in pred_cols:
        pred = str(row[col])
        hyp = [pred]  # list of hypothesis strings

        # Compute BLEU score using SacreBLEU
        bleu = sacrebleu.corpus_bleu(hyp, ref).score

        row_scores[f"{col}_BLEU"] = bleu

    results.append(row_scores)

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# === Step 3: Save to CSV ===
results_df.to_csv("Lora_BlueScore.csv", index=False)

print("Done! Saved to output_bleu_scores.csv")
```
