```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util

# all-mpnet-base-v2
# all-MiniLM-L6-v2
model = SentenceTransformer('all-MiniLM-L6-v2')

def semantic_factual_score(model_answer, ground_truth):
    emb_model = model.encode(model_answer, convert_to_tensor=True)
    emb_gt = model.encode(ground_truth, convert_to_tensor=True)
    score = util.cos_sim(emb_model, emb_gt).item()
    return score    # 0 - 1

# Load CSV
df = pd.read_csv("Result_Baseline.csv")

# Columns
trained_col = "Trained Response"   # second column
compare_cols = df.columns[3:]      # from 4th column onward

# New dataframe to store scores
score_df = pd.DataFrame()
score_df["Index"] = df.index
score_df[trained_col] = df[trained_col]

# Compute similarity scores
for col in compare_cols:
    score_df[col + "_score"] = df.apply(
        lambda row: semantic_factual_score(str(row[trained_col]), str(row[col]))
        if pd.notna(row[col]) else None,
        axis=1
    )

# Save output
score_df.to_csv("Baseline_MiniLM.csv", index=False)
print("Saved semantic_scores.csv")
```
